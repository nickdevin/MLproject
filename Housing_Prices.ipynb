## Random forest model

```python
from sklearn import ensemble
rf_housing = ensemble.RandomForestRegressor()
rf_housing.set_params(random_state = 42, n_estimators=500, max_features = 9)
```

```python
# dummify categorical features for random forest

# train = pd.get_dummies(train)
train.sample(15)
```

```python
rf_housing.fit(train.drop('SalePrice', axis = 1),train.loc[:,'SalePrice'])
```

```python
feature_importance = list(zip(train.columns[:-2], rf_housing.feature_importances_))
dtype = [('feature', 'S10'), ('importance', 'float')]
feature_importance = np.array(feature_importance, dtype=dtype)
feature_sort = np.sort(feature_importance, order='importance')[::-1]
name, score = zip(*list(feature_sort))
pd.DataFrame({'name':name,'score':score})[:15].plot.bar(x='name', y='score')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train.drop('SalePrice', axis = 1),train.loc[:,'SalePrice'], random_state = 0, test_size = 0.2)
```
